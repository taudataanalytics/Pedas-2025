```python
import warnings; warnings.simplefilter('ignore')
import pandas as pd, numpy as np
from sklearn import preprocessing
from tqdm import tqdm

fData = "data/dataTransaksi_PeDaS_2024.zip"
label_encoder = preprocessing.LabelEncoder() 

df = pd.read_csv(fData, compression='zip')
df["creationdate"] = pd.to_datetime(df['creationdate'])
df['trx_code']= label_encoder.fit_transform(df['trx_code'])
df['rek_code']= label_encoder.fit_transform(df['rek_code']) 
df['rek']= label_encoder.fit_transform(df['rek'])
df['type']= label_encoder.fit_transform(df['type'])
df['Occurence'] = df.groupby('rek_code')['rek_code'].transform('size')
print(len(df.trx_code.unique()), len(df.trx_id.unique()), len(df.rek_code.unique()))
df.info()
```

```python
df.head()
```

```python
df = df[df.Occurence >=350 ]
print(len(df.trx_code.unique()), len(df.trx_id.unique()), len(df.rek_code.unique()))
df.head()
```

```python
df_test = df.drop(df.index)

for i, r in tqdm(enumerate(df.rek_code.unique())):
    dfR = df.loc[df['rek_code'] == r].sort_values(by='creationdate')
    last = dfR.tail(1)
    df_test = pd.concat([df_test, last])
    df = df.drop(last.index[0])

df.drop(['Occurence'], axis=1, inplace=True)
df_test_peserta = df_test.drop(['Occurence','amount'], axis=1)
df_test_peserta['balance'] = None
df_test_peserta['type'] = None
```
